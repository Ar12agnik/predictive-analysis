```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
```

```python
dataset = pd.read_csv("/content/Kolkata.csv")
dataset
```

```python


# Assuming your DataFrame is named 'dataset'
fields = ['MaintenanceStaff', 'Gymnasium', 'SwimmingPool', 'LandscapedGardens',
       'JoggingTrack', 'RainWaterHarvesting', 'IndoorGames', 'ShoppingMall',
       'Intercom', 'SportsFacility', 'ATM', 'ClubHouse', 'School',
       '24X7Security', 'PowerBackup', 'CarParking', 'StaffQuarter',
       'Cafeteria', 'MultipurposeRoom', 'Hospital', 'WashingMachine',
       'Gasconnection', 'AC', 'Wifi', 'Children\'splayarea', 'LiftAvailable',
       'BED', 'VaastuCompliant', 'Microwave', 'GolfCourse', 'TV',
       'DiningTable', 'Sofa', 'Wardrobe', 'Refrigerator']

# Count the number of '9' values in each field
counts = {}
for field in fields:
    try:
        counts[field] = dataset[dataset[field] == 9][field].count()
    except KeyError:
        print(f"Warning: Field '{field}' not found in the dataset.")
        counts[field] = 0  # Or handle the missing field differently

# Create the bar plot
plt.figure(figsize=(15, 6))  # Adjust figure size for better readability
plt.bar(counts.keys(), counts.values())
plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility
plt.xlabel("Fields")
plt.ylabel("Number of '9' values")
plt.title("Number of '9' values in each field")
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
newdata = dataset[['Price', 'Area', 'Location', 'No. of Bedrooms']]
```

```python
newdata
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
newdata.columns
```

```python
x=newdata[['Area', 'No. of Bedrooms']][:100]
y=newdata['Price'][:100]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
for a, b in zip(y_pred, y_test.values):
    print(f"Predicted: {a:.2f}  Actual: {b}")
```

```python
mse = mean_squared_error(y_test,y_pred)
mse
```

```python
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
```

```python
newdata.columns
```

```python
#x_train, x_test, y_train, y_test
scaller = StandardScaler()
x_transform_train = scaller.fit_transform(x_train)
x__transform_test = scaller.transform(x_test)
```

```python
svr =  SVR(kernel='rbf', C=10000, epsilon=500, gamma='scale')
svr.fit(x_transform_train, y_train)
```

```python
y_pred = svr.predict(x__transform_test)
mse=mean_squared_error(y_test, y_pred)
mse
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"R² Score: {r2}")

```
